```python
def gcd(a,b):
    if not isinstance(a,int) or not isinstance(b, int):
        raise TypeError("input must be digit.")
        
    while b == 0:
        return a 
    return gcd(b, a % b)

if __name__ == "__main__":
    try:
        a = 4
        b = 16
        if a == "" or b== "":
            print("empty")
        else:
            print(gcd(a,b))
    except (TypeError , ValueError) as e:
        print("invalid output")
    except Exception as e:
        print("error",e)
```
